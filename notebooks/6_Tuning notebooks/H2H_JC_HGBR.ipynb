Set parameters

```python
import JiaoCheng_H2H
```

```python
jiaocheng = JiaoCheng_H2H.JiaoCheng()
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor as HGBR
```

```python
jiaocheng.read_in_model(HGBR, 'Regression')
```

```python
jiaocheng.read_in_data(train_x, train_y, val_x, val_y, test_x, test_y)
```

```python
parameter_choices = {
    'max_depth': (3, 6, 12, 24, 48, 96, 192),
    'max_bins': (4, 16, 64, 255),
    'interaction_cst': ('no_interactions', 'pairwise'),
    'learning_rate': (0.00000001, 0.000001, 0.0001, 0.01, 0.1),
    'l2_regularization': (0.0001, 0.001, 0.01, 0.1, 1, 10)

}
```

```python
jiaocheng.set_hyperparameters(parameter_choices)
```

```python
jiaocheng.set_non_tuneable_hyperparameters({'random_state': 19260817})
```

```python
jiaocheng.set_features(ningxiang_output)
```

```python
jiaocheng.set_tuning_order(['features', 'max_depth', 'max_bins', 'interaction_cst', 'l2_regularization', 'learning_rate'])
```

```python
jiaocheng.set_hyperparameter_default_values({
    'features': 0,
    'max_depth': 12,
    'max_bins': 255,
    'interaction_cst': 'pairwise',
    'learning_rate': 0.1,
    'l2_regularization': 0.01}
  )
```
